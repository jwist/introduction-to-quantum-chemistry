### the proper coordinate system

We first write down the wave function with its degree of freedom

$$
\Psi(X_N, Y_N ,Z_N ,X_1 ,Y_1 ,Z_1 ,X_2 ,Y_2 ,Z_2)
$$

We consider the center of mass as origin for our system. In that case our function recast to:

$$
\varPsi(x_{1}, y_{1},z_{1},x_{2},y_{2},z_{2})
$$

In atomic units and in the coordinates of the center of mass the *Hamiltonian* is:

$$
\hat{\mathcal{H}}=-\frac{1}{2}\nabla_1^2-\frac{1}{2}\nabla_2^2-\frac{2}{\sqrt{x_1^2+y_1^2+z_1^2}}-\frac{2}{\sqrt{x_2^2+y_2^2+z_2^2}}
+\frac{1}{\sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2}}
$$

Remember... the cartesian system is not really good for spherical symmetries

$$
\varPsi(x_{1},y_{1}, z_{1}, x_{2}, y_{2}, z_{2})\equiv\varPsi(r_1, \theta_1, \phi_1, r_2, \theta_2, \phi_2)
$$


### the Schr√∂dinger equation

$$
\varPsi(\vec{r}_1,\vec{r}_2)
$$

$$
\hat{\mathcal{H}}=-\frac{1}{2}\nabla_1^2-\frac{1}{2}\nabla_2^2-\frac{2}{\vec{r}_1}-\frac{2}{\vec{r}_2}+\frac{1}{r_{12}}
$$

$$
\hat{\mathcal{H}}\varPsi(\vec{r}_1,\vec{r}_2)=E_e\varPsi(\vec{r}_1,\vec{r}_2)
$$

$r^{-1}_{12}$ makes the separation of variable impossible!




### first approximation 

as a first approximation we can simply drop the repulsion term. In this case we can write our *Hamiltonian* as the sum:

$$
\hat{H}=\hat{h}(\vec{r}_1)+\hat{h}(\vec{r}_2)
$$
$$
\hat{h}(\vec{r}_i)=-\frac{1}{2}\nabla_i^2-\frac{2}{\vec{r}_i}
$$

And the equation to be solved is:

$$
\hat{H}\psi(\vec{r}_1,\vec{r}_2)=W\psi(\vec{r}_1,\vec{r}_2)
$$

We only have to assume $\psi(\vec{r}_1,\vec{r}_2)=\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)$ in order to proceed with the separation of variables.




### the separation of variables

$$
\hat{H}\psi(\vec{r}_1,\vec{r}_2)=W\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)
$$
$$
\Bigl[\hat{h}(\vec{r}_1)+\hat{h}(\vec{r}_2)\Bigr]\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)=W\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)
$$

$$
\hat{h}(\vec{r}_1)\phi_1(\vec{r}_1)=\epsilon_1\phi_1(\vec{r}_1)
$$
$$
\hat{h}(\vec{r}_2)\phi_2(\vec{r}_2)=\epsilon_2\phi_2(\vec{r}_2)
$$

$$
W=\epsilon_1+\epsilon_2
$$

Ignoring the repulsion term we can factor the wave function, that is we consider the total wavefunction as the product of two *monoelectronic* functions, or 2 *orbitals*. Since we already know theses functions we can readily solve the problem:

$$
\phi_i(\vec{r}_i)\in \{ 1s(\vec{r}_i), 2s(\vec{r}_i), 2p_x(\vec{r}_i), 2p_y(\vec{r}_i), 2p_z(\vec{r}_i), 3s(\vec{r}_i), \ldots \}
$$

So we have to find two functions that give the lowest possible energy:

$$
\psi(\vec{r}_1,\vec{r}_2)=\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)=1s_1(\vec{r}_1)1s_2(\vec{r}_2)
$$

and the energy (a.u.) is:

$$
W = \epsilon_1 + \epsilon_2 =\frac{-Z^2}{2n_1^2} + \frac{-Z^2}{2n_2^2}= -4 \quad hartrees
$$

## variational teorem

Because we know the correct *Hamiltonian* but we don't know how to find the eignenfunction, we can use the following equation to find an approximation to the energy:

$$
\langle E \rangle= \frac {\iint\limits\psi^*(\vec{r}_1,\vec{r}_2)\hat{\mathcal{H}}\,\psi(\vec{r}_1,\vec{r}_2)d\tau_1\,d\tau_2}{ \iint\limits\psi^*(\vec{r}_1,\vec{r}_2)\psi(\vec{r}_1,\vec{r}_2)d\tau_1\,d\tau_2} 
$$

**NOTE:** in order to obtain an approximation for the energy we need to provide *functions*. Therefore the importance of orbitals

Looking at the denominator: 

$$
\iint\limits1s_1^2(\vec{r}_1)1s_2^2(\vec{r}_2)d\tau_1\,d\tau_2=\int\limits|1s_1(\vec{r}_1)|^2d\tau_1\int\limits|1s_2(\vec{r}_2)|^2d\tau_2 \equiv 1
$$

$$
\langle E \rangle =\iint\limits 1s_1(\vec{r}_1)1s_2(\vec{r}_2) \hat{\mathcal{H}}\,1s_1(\vec{r}_1)1s_2(\vec{r}_2)d\tau_1\,d\tau_2
$$
$$
\langle E \rangle =\iint\limits 1s_1(\vec{r}_1)1s_2(\vec{r}_2)\Bigl[\hat{h}(\vec{r}_1)+\hat{h}(\vec{r}_2)+r_{12}^{-1}\Bigr]1s_1(\vec{r}_1)1s_2(\vec{r}_2)d\tau_1\,d\tau_2
$$

$$
\langle E \rangle=\iint\limits 1s_1(\vec{r}_1)1s_2(\vec{r}_2)\hat{h}(\vec{r}_1)1s_1(\vec{r}_1)1s_2(\vec{r}_2)d\tau_1\,d\tau_2 +  \\
\iint\limits 1s_1(\vec{r}_1)1s_2(\vec{r}_2)\hat{h}(\vec{r}_2)1s_1(\vec{r}_1)1s_2(\vec{r}_2)d\tau_1\,d\tau_2 + \\
\iint\limits 1s_1(\vec{r}_1)1s_2(\vec{r}_2)\,r_{12}^{-1}1s_1(\vec{r}_1)1s_2(\vec{r}_2)d\tau_1\,d\tau_2
$$

We already solved the first two lines and remember that they are equal to $\epsilon_1$ and $\epsilon_2$. We only have to solve the third line.

$$\langle E \rangle=\epsilon_1\,+ \epsilon_2\, + \frac{5}{8}\,Z=-2-2+5/4=-2.75 \quad hartrees$$

Let's look at the sign of the contributions. The monoelectronic parts are negative and thus contribute to the stabilization of the atom, while the repulsion term (Coulomb integral) is positive.

## the symmetry and the anti-symmetry

For polyelectronic atoms we obtain wavefunctions that depends on degrees of freedom for each particle. Remember the trial wavefunction $\psi(\vec{r}_1,\vec{r}_2)=\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)$. 

We might reckon however that the results of the calculation cannot change because we swap the two electrons. **that is we must require that the electrons are indistinguishable**



Let's consider an operator $\hat{P}_{ij}$ whose effect is to swap two particles $i$ and $j$, we can write the eignen equation:

$$\hat{P}_{ij}w(r_1,\ldots, r_i,r_j,\ldots)=c w(r_1,\ldots, r_j,r_i,\ldots )$$

If we apply a second time the operator:

$$\hat{P}_{ij}\hat{P}_{ij}w(r_1,\ldots, r_i,r_j,\ldots)=c \hat{P}_{ij}w(r_1,\ldots, r_j,r_i,\ldots )$$

But by definition $\hat{P}_{ij}\hat{P}_{ij}=\mathbb{1}$ so

$$w(r_1,\ldots, r_i,r_j,\ldots)=c^2 w(r_1,\ldots, r_i,r_j,\ldots )$$

and $c$ should be $\pm 1$ meaning that **swapping** two electron should leave the function untouched or multiplied by -1. Thus the wavefunction must be **symmetric** or **anti-symmetric** with respect to swapping particles



The requirement that electrons are indistinguishable comes from the fact that we cannot measure repeatidly the postion of $e_1$ and $e_2$ to obtain an expected value. Indeed, suppose that we could measure the position of $e_1$ and $e_2$, when we repeat the measurement, how can we be sure that we get the label correctly? we **cannot** be sure, because we **cannot** distinguish the electrons!

## spin-orbital

Electrons are fermions and fermions are described by anti-symmetric wavefunctions. But our spatial orbital $\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)$ is symmetric, this means that the spin wavefunction must be anti-symmetric.

For instance:

$$\alpha(1)\alpha(2)=\alpha(2)\alpha(1)$$
$$\beta(1)\beta(2)=\beta(2)\beta(1)$$

are symmetric functions

and

$$\alpha(1)\beta(2)$$
$$\beta(1)\alpha(2)$$

are neither symmetric nor anti-symmetric. But we can combine them to obtain 

$$\frac{1}{\sqrt{2}}(\alpha(1)\beta(2) + \beta(1)\alpha(2))$$
$$\frac{1}{\sqrt{2}}(\alpha(1)\beta(2) - \beta(1)\alpha(2))$$

the first one is symmetric, while the second one is anti-symmetric. So the spin-orbital is:

$$\sigma(\vec{r}_1,\vec{r}_2,s_1,s_2)=\phi_1(\vec{r}_1)\phi_2(\vec{r}_2)\frac{1}{\sqrt{2}}[(\alpha(1)\beta(2) - \beta(1)\alpha(2)]$$


## effective charge

\begin{equation}
\psi(\vec{r}_1,\vec{r}_2)=1s(\vec{r}_1)1s(\vec{r}_2)=\tfrac{Z^2}{\pi}e^{-Z(r_1+r_2)}
\end{equation}

\begin{equation}
\begin{split}
\psi^{modificada}=\psi'(\vec{r}_1,\vec{r}_2;\gamma)=&\frac{(\gamma Z)^2}{\pi}e^{-\gamma Z(r_1+r_2)}=\\
=&K'e^{-Z^*(r_1+r_2)}=1s'(\vec{r_1})1s'(\vec{r_2}),
\end{split}
\end{equation}

\begin{equation}
E(Z^{*})=\frac{\bigl<\psi^{mod}|\mathcal{H}|\psi^{mod}\bigr>}{\bigl<\psi^{mod}|\psi^{mod}\bigr>}=Z^{*2}-2ZZ^{*}+\frac{5}{8}Z^{*} 
\end{equation}



```python
import sympy as sp
#import numpy as np
Z = 2
z = sp.Symbol('z')
y = z**2 - 2*Z*z + (5/8)*z
yprime = y.diff(z)
print('derivative: ' + str(yprime))
sol = sp.solve(yprime)
print('solution (Z*): ' + str(sol[0].round(4)))
print('shielding constant Z-Z*: ')
print(sp.nsimplify(Z-sol[0]))
```

\begin{equation}
\frac{dE}{dZ^*}=2Z^*-2Z+\frac{5}{8}=0.
\end{equation}


The optimal value is $Z^*=Z-\frac{5}{16}$ and by replacing $Z^*$ we obtain an energy of -2,8477 hartrees, less that the -2,75 hartrees that we got initially, and closer to the exeprimental -2.9037.

\begin{equation}
\begin{split}
\psi^{mejorada}=\psi'(\vec{r}_1,\vec{r}_2;\gamma)=&\tfrac{(\gamma Z)^2}{\pi}e^{-\gamma Z(r_1+r_2)}=\\
=&K'e^{-\frac{27}{32}Z(r_1+r_2)}=K'e^{-\frac{27}{16}(r_1+r_2)}
\end{split}
\end{equation}

and $\sigma=Z-Z^*$ is referred to as the shielding constant.

```python
# we can no evaluate the energy at the optimal value of Z*
print('our best approximation for energy is ' + str( y.evalf(subs={z:sol[0]}).round(3)) )
```

## Koopman's theorem

The koopman's theorem states that the **binding energy (BE)** is equal to 

$$
BE = E_{final}(n-1) - E_{initial}(n)
$$

So BE can be computed within the Hartree Fock theory assuming that
 - no orbital relaxation occurs
 - electronic correlation effects are weak

In other word that K's theorem says is that the wavefunctions of the neutral atom are valid to describe the ion.
